```python
#загрузка необходимых библиотек
import pandas as pd
import scipy.stats as sts
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#генерация выборки случайных величин гамма-распределения
a=3.
gamma_rv=sts.gamma(a)
sample=gamma_rv.rvs(1000)
```

```python
#построение гистограммы выборки и графика теоретической плотности гамма-распределения
plt.hist(sample, density=True)
plt.ylabel('fraction of samples')
plt.xlabel('x')

x=np.linspace(0,12,1000)
pdf=gamma_rv.pdf(x)
plt.plot(x,pdf, label='theoretical pdf', alpha=0.7)
plt.legend()
```

```python
#значение параметров гамма-распеределения (мат ожидание и дисперсия)
E=a
D=a
```

```python
#функция для генерации выборочных средних из 1000 выборок объема n
def average_sample(n):
    average=list()
    for i in xrange(1000):
        sample=gamma_rv.rvs(n)
        average.append(np.mean(sample))
    return average
```

```python
#построение гистограммы распределения выборочных средних из выборок объема n=5
plt.hist(average_sample(5), density=True)
plt.ylabel('fraction of average samples')
plt.xlabel ('x')
plt.xlim(1,6)
# построение графика теоретической плотности нормалньного распределения 
# приближающего распределние выборочных средних из выборок объема n=5
norm_rv=sts.norm(E, (D/5)**(0.5)) #в скобках вычисляются параметры 𝜇  и 𝜎 приближаюего нормального распределения по ЦПТ
x=np.linspace(1,6,1000)
pdf=norm_rv.pdf(x)
plt.plot(x, pdf, label='approx by normal  pdf', alpha=0.7)
plt.legend()
```

```python
#построение гистограммы распределения выборочных средних из выборок объема n=10
plt.hist(average_sample(10), density=True)
plt.ylabel('fraction of average samples')
plt.xlabel ('x')
plt.xlim(1,6)
# построение графика теоретической плотности нормалньного распределения 
# приближающего распределние выборочных средних из выборок объема n=10
norm_rv=sts.norm(E, (D/10)**(0.5))#в скобках вычисляются параметры 𝜇  и 𝜎 приближаюего нормального распределения по ЦПТ
x=np.linspace(1,6,1000)
pdf=norm_rv.pdf(x)
plt.plot(x, pdf, label='approx by normal  pdf', alpha=0.7)
plt.legend()
```

```python
#построение гистограммы распределения выборочных средних из выборок объема n=50
plt.hist(average_sample(50), density=True)
plt.ylabel('fraction of average samples')
plt.xlabel ('x')
plt.xlim(1,6)
# построение графика теоретической плотности нормалньного распределения 
# приближающего распределние выборочных средних из выборок объема n=50
norm_rv=sts.norm(E, (D/50)**(0.5))#в скобках вычисляются параметры 𝜇  и 𝜎 приближаюего нормального распределения по ЦПТ
x=np.linspace(1,6,1000)
pdf=norm_rv.pdf(x)
plt.plot(x, pdf, label='approx by normal  pdf', alpha=0.7)
plt.legend()
```

   Разница полученных распределений при различных n в уменьшении разброса выборочных средних от наиболее частого значения, а также увеличивается доля этого значения.
