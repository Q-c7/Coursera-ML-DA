```python
#–∑–∞–≥—Ä—É–∑–∫–∞ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –±–∏–±–ª–∏–æ—Ç–µ–∫
import pandas as pd
import scipy.stats as sts
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#–≥–µ–Ω–µ—Ä–∞—Ü–∏—è –≤—ã–±–æ—Ä–∫–∏ —Å–ª—É—á–∞–π–Ω—ã—Ö –≤–µ–ª–∏—á–∏–Ω –≥–∞–º–º–∞-—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
a=3.
gamma_rv=sts.gamma(a)
sample=gamma_rv.rvs(1000)
```

```python
#–ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –≤—ã–±–æ—Ä–∫–∏ –∏ –≥—Ä–∞—Ñ–∏–∫–∞ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≥–∞–º–º–∞-—Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
plt.hist(sample, density=True)
plt.ylabel('fraction of samples')
plt.xlabel('x')

x=np.linspace(0,12,1000)
pdf=gamma_rv.pdf(x)
plt.plot(x,pdf, label='theoretical pdf', alpha=0.7)
plt.legend()
```

```python
#–∑–Ω–∞—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –≥–∞–º–º–∞-—Ä–∞—Å–ø–µ—Ä–µ–¥–µ–ª–µ–Ω–∏—è (–º–∞—Ç –æ–∂–∏–¥–∞–Ω–∏–µ –∏ –¥–∏—Å–ø–µ—Ä—Å–∏—è)
E=a
D=a
```

```python
#—Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ –≤—ã–±–æ—Ä–æ—á–Ω—ã—Ö —Å—Ä–µ–¥–Ω–∏—Ö –∏–∑ 1000 –≤—ã–±–æ—Ä–æ–∫ –æ–±—ä–µ–º–∞ n
def average_sample(n):
    average=list()
    for i in xrange(1000):
        sample=gamma_rv.rvs(n)
        average.append(np.mean(sample))
    return average
```

```python
#–ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤—ã–±–æ—Ä–æ—á–Ω—ã—Ö —Å—Ä–µ–¥–Ω–∏—Ö –∏–∑ –≤—ã–±–æ—Ä–æ–∫ –æ–±—ä–µ–º–∞ n=5
plt.hist(average_sample(5), density=True)
plt.ylabel('fraction of average samples')
plt.xlabel ('x')
plt.xlim(1,6)
# –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –Ω–æ—Ä–º–∞–ª–Ω—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è 
# –ø—Ä–∏–±–ª–∏–∂–∞—é—â–µ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–Ω–∏–µ –≤—ã–±–æ—Ä–æ—á–Ω—ã—Ö —Å—Ä–µ–¥–Ω–∏—Ö –∏–∑ –≤—ã–±–æ—Ä–æ–∫ –æ–±—ä–µ–º–∞ n=5
norm_rv=sts.norm(E, (D/5)**(0.5)) #–≤ —Å–∫–æ–±–∫–∞—Ö –≤—ã—á–∏—Å–ª—è—é—Ç—Å—è –ø–∞—Ä–∞–º–µ—Ç—Ä—ã ùúá  –∏ ùúé –ø—Ä–∏–±–ª–∏–∂–∞—é–µ–≥–æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ –¶–ü–¢
x=np.linspace(1,6,1000)
pdf=norm_rv.pdf(x)
plt.plot(x, pdf, label='approx by normal  pdf', alpha=0.7)
plt.legend()
```

```python
#–ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤—ã–±–æ—Ä–æ—á–Ω—ã—Ö —Å—Ä–µ–¥–Ω–∏—Ö –∏–∑ –≤—ã–±–æ—Ä–æ–∫ –æ–±—ä–µ–º–∞ n=10
plt.hist(average_sample(10), density=True)
plt.ylabel('fraction of average samples')
plt.xlabel ('x')
plt.xlim(1,6)
# –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –Ω–æ—Ä–º–∞–ª–Ω—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è 
# –ø—Ä–∏–±–ª–∏–∂–∞—é—â–µ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–Ω–∏–µ –≤—ã–±–æ—Ä–æ—á–Ω—ã—Ö —Å—Ä–µ–¥–Ω–∏—Ö –∏–∑ –≤—ã–±–æ—Ä–æ–∫ –æ–±—ä–µ–º–∞ n=10
norm_rv=sts.norm(E, (D/10)**(0.5))#–≤ —Å–∫–æ–±–∫–∞—Ö –≤—ã—á–∏—Å–ª—è—é—Ç—Å—è –ø–∞—Ä–∞–º–µ—Ç—Ä—ã ùúá  –∏ ùúé –ø—Ä–∏–±–ª–∏–∂–∞—é–µ–≥–æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ –¶–ü–¢
x=np.linspace(1,6,1000)
pdf=norm_rv.pdf(x)
plt.plot(x, pdf, label='approx by normal  pdf', alpha=0.7)
plt.legend()
```

```python
#–ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –≤—ã–±–æ—Ä–æ—á–Ω—ã—Ö —Å—Ä–µ–¥–Ω–∏—Ö –∏–∑ –≤—ã–±–æ—Ä–æ–∫ –æ–±—ä–µ–º–∞ n=50
plt.hist(average_sample(50), density=True)
plt.ylabel('fraction of average samples')
plt.xlabel ('x')
plt.xlim(1,6)
# –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏–µ –≥—Ä–∞—Ñ–∏–∫–∞ —Ç–µ–æ—Ä–µ—Ç–∏—á–µ—Å–∫–æ–π –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –Ω–æ—Ä–º–∞–ª–Ω—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è 
# –ø—Ä–∏–±–ª–∏–∂–∞—é—â–µ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–Ω–∏–µ –≤—ã–±–æ—Ä–æ—á–Ω—ã—Ö —Å—Ä–µ–¥–Ω–∏—Ö –∏–∑ –≤—ã–±–æ—Ä–æ–∫ –æ–±—ä–µ–º–∞ n=50
norm_rv=sts.norm(E, (D/50)**(0.5))#–≤ —Å–∫–æ–±–∫–∞—Ö –≤—ã—á–∏—Å–ª—è—é—Ç—Å—è –ø–∞—Ä–∞–º–µ—Ç—Ä—ã ùúá  –∏ ùúé –ø—Ä–∏–±–ª–∏–∂–∞—é–µ–≥–æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ –¶–ü–¢
x=np.linspace(1,6,1000)
pdf=norm_rv.pdf(x)
plt.plot(x, pdf, label='approx by normal  pdf', alpha=0.7)
plt.legend()
```
