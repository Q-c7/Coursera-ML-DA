```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

## Часть 1: генерация выборки

Здесь я взял экспоненциальное распределение; параметр $\lambda$ = 3.
Генерирую так же, как в готовых шаблонах ipynb Курсеры.

```python
lam = 3.
# зададим экспоненциальное распределение с параметром lambda = 3 (см. википедию), в питоне это задаётся через scale
exp_rv = sts.expon(loc = 0, scale = 1 / lam)

# сгенерируем 1000 значений
hist1_val = exp_rv.rvs(size = 1000)
```

```python
n_bin = 100
x_max = 2

# строим полученную гистограмму
plt.hist(hist1_val, density = True, range = (0, x_max), bins = n_bin, label = 'generated histogram')
plt.ylabel('fraction of samples, n=1000')
plt.xlabel('$x$')

# и распределение для сравнения
x = np.linspace(0, x_max, n_bin)
plt.plot(x, exp_rv.pdf(x), label = 'theoretical pdf', linewidth = 3, c = 'red')

plt.legend(loc='upper right')
plt.title('task 1: one set with n = 1000 & expon pdf')
```

## Часть 2: генерация 1000 выборок, распределение среднего, проверка ЦПТ

Аналогично ч. 1, будем генерировать уже 1000 выборок, чтобы получить по 1000 средних значений для гистограммы.
Прошу прощения за быдлокод :) скорее всего, гистограмму средних значений можно получить быстрее, но это сейчас не особо важно

```python
# Вначале зададим теор. параметры нормального распределения; именно к этой функции мы и хотим приблизить наши гистограммы
# M[X] = 1/lam (source:wiki)
mu = 1/lam
# D[X] = 1/lam^2 => sigma = 1/lam
sigma = 1/lam
```

```python
def gen_and_draw(n_stat, n_bin, x_max):
    
    # Получим нормальное распределение
    # Не забываем поправить дисперсию на n, как утверждает ЦПТ!
    norm_rv = sts.norm(loc=mu, scale=sigma/math.sqrt(n_stat))
    
    # Теперь сделаем данные для гистограммы
    hist_mean = np.zeros(1000)
    # Будем 1000 раз генерировать выборку длиной n и в лоб считать её среднее
    for i in np.arange(1000):
        #Генерируем выборку
        cur_hist_val = exp_rv.rvs(size = n_stat)
        #Считаем среднее
        cur_avg = sum(cur_hist_val) / len(cur_hist_val) 
        hist_mean[i] = cur_avg
       
    # Строим гистограмму средних значений
    plt.hist(hist_mean, density=True, bins=n_bin, label = 'mean values histogram')

    #Гауссиану сделаем в 20 раз более гладкую, чем гистограмму
    x = np.linspace(0, x_max, n_bin * 20) 

    # Строим pdf
    plt.plot(x, norm_rv.pdf(x), label = 'gaussian pdf for comparison', linewidth = 2, c = 'red')

    plt.ylabel('fraction of events; total n = 1000')
    plt.xlabel('mean value')

    plt.legend(loc='upper right')
    
```

```python
gen_and_draw(5, 100, 1)
plt.title('task 2.1: mean values histogram, n = 5')
```

Дальше аналогично; n = 10 не сильно улучшает ситуацию. Скорее всего, причина - скошенность экспоненциального расппределения.

```python
gen_and_draw(10, 100, 1)
plt.title('task 2.2: mean values histogram, n = 10')
```

n = 50 уже даёт не самую плохую картинку

```python
gen_and_draw(50, 100, 1)
plt.title('task 2.3: mean values histogram, n = 50')
```

Вывод: экспоненциальное распределение не является исключением; для него ЦПТ тоже отлично сработала, несмотря на скошенность. Даже для n = 50 мы уже имеем достаточно хорошую картинку, как и предсказывалось.

```python
gen_and_draw(150, 100, 1)
plt.title('bonus: mean values histogram, n = 150')
```

Bonus: видим, что при увеличении n до 150 всё стало еще лучше: на 100 бинов статистики косячат около 5. Если уменьшить количество бинов, то картинка будет более гладкая и всё будет еще красивее.

```python
gen_and_draw(150, 20, 1)
plt.title('bonus: mean values histogram, n = 150, less bins')
```
