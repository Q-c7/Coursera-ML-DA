```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

## Экспоненциальное распределение.

Для примера рассмотрим стандартизированное экспоненциальное случайное распределение с параметром $\lambda = 1$.
Плотность вероятности для экспоненциального распределения будет иметь следующий вид:

$
f_X(x) = \begin{cases}
\lambda \,e^{-\lambda x} , x \ge 0, \\
0 , x < 0.
\end{cases}.
$

Для такого распределения геренируем случайную выборку размером 1000.

```python
#exponential
from scipy.stats import expon
ex = expon.rvs(size=1000)
```

```python
rv = sts.expon() 
ext = np.linspace(0, 8, 1000)
```

```python
plt.hist(ex, density = True, label='Histogram')
#plt.plot(pdf, label='Theoretical PDF', alpha=0.5)
plt.plot(ext, expon.pdf(ext),
         lw=5, alpha=0.6, label='Theoretical PDF')
plt.ylabel('f(x)')
plt.xlabel('x')
plt.legend()
```

## Распределение Выборочного Среднего (ЦПТ)


В случае со стандартизированным экспоненциальным распределением (как выше), матожидание равно  $\mathbb{E}X = \lambda^{-1}\, = 1$ , дисперсия $\mathbb{D}X = \lambda^{-2}\, = 1$.
Тогда по ЦПТ, распределение выборочных средних при $\lambda = 1$ можно описать нормальным распределением $N(\mu = 1, \sigma^2 = \frac{1}{n})$.

Для того, чтобы работать с распределением распределением выборочных средних, удобно написать функцию, генерирующую N (1000) выборок размером n (5, 10, 50), оценивающую их среднее и описывает распределение средних согласно ЦПТ.

```python
N = 1000
n_lis = [5,10,50]
```

```python
def stand_expon_pdf(N, n, mean = 1): 
    avg_lis = []
    for i in range(N):
        x = rv.rvs(n)

        avg = np.mean(x) #среднее по выборке
        avg_lis.append(avg)
    
    norm_rv = sts.norm(loc=mean, scale=(1./n) ** 0.5) #loc = mean or mu, scale = sigma теоретического распределения
    xn = np.linspace(0,4,N)
    n_pdf = norm_rv.pdf(xn)

    
    plt.hist(avg_lis, density = True, 
             label='Histogram of Averages')
    plt.plot(xn, n_pdf,
         lw=5, alpha=0.6, label='Theoretical PDF')
    plt.legend()
    
    plt.xlabel('Среднее по выборке')
    plt.ylabel('Оценка PDF')
    plt.title('Размер выборки (n) = %d' % n)
    plt.show()
```

```python
for n in n_lis:
    stand_expon_pdf(N, n, 1)
```

## Вывод
Так как изначальное распределение скошено, то при использовании ЦПТ для аппроксимации распределения выборочных средних нормальным распределением требуется большее число выборок (согласно практике, больше 30). Смотря на графики выше, понятно, что распределение изменяется: "центруется" в районе среднего значения = 1 и все более напоминает нормальное.

```python
stand_expon_pdf(N, 100, 1)
```

```python
stand_expon_pdf(N, 200, 1)
```
