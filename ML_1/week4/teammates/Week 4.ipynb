```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy.stats as sts
%matplotlib inline
```

```python
# выборка объёма 1000 из биномиального распределения, m = 15, k = 0.4
m = 15
k = 0.4
binomial_rv = sts.binom(m, k)
sample = binomial_rv.rvs(1000)
E = m*k #мат ожидание
D = m*k*(1-k) #дисперсия
```

```python
# гистограмма выборки
plt.hist(sample, label='hist',  normed=True)
plt.xlabel('$x$')

# теоретическая плотность распределения:
x = np.linspace(0,20,21)
pmf = binomial_rv.pmf(x)
plt.plot(x, pmf, label='theoretical pmf', alpha=1)
plt.legend()
plt.ylabel('$f(x)/number of samples$')
plt.xlabel('$x$')
```

```python
# генерация 1000 выборок объема n = 5 
n = 5
v = 1000
sample5 = np.array([]) # определение массива выборок
for i in np.arange(0,v,1):
    sample5i = binomial_rv.rvs(n)
    avg5i = sum(sample5i)*1./n # определение выборочного среднего
    sample5 = np.append(sample5, avg5i) # добавление в массив новое значение выборочного среднего
```

```python
# параметры распределения, описывающие массив 1000 выборочных средних при n=5
D5 = D/n # дисперсия для распределения, приближающего выборку средних
sigma5 = math.sqrt(D5) # сигма для распределения выборки средних
norm5 = sts.norm(E, sigma5) # распределение
x = np.linspace(0,20,100)
pdf5 = norm5.pdf(x) # плотность распределения
# гистограмма массива средних при n=5 и плотность описывающая этот массив нормального распределения
plt.plot(x, pdf5, label='PDF(n=5)')
plt.hist(sample5, normed=True)
plt.ylabel('number of samples')
plt.xlabel('$x$')
plt.legend(loc='best')
```
