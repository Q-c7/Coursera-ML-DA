```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('weights_heights.csv', index_col='Index')
```

```python
data[:5]
```

```python
data.plot(y='Weight', kind='hist', 
           color='green',  title='Weight (inch.) distribution')
```

```python
def make_bmi(height_inch, weight_pound):
    METER_TO_INCH, KILO_TO_POUND = 39.37, 2.20462
    return (weight_pound / KILO_TO_POUND) / \
           (height_inch / METER_TO_INCH) ** 2
```

```python
data['BMI'] = data.apply(lambda row: make_bmi(row['Height'], 
                                              row['Weight']), axis=1)
```

```python
g = sns.pairplot(data)
```

```python
def weight_category(weight):
    return 1 if weight < 120 else 3 if weight >= 150 else 2
data['weight_cat'] = data['Weight'].apply(weight_category)
print(data.head())
```

```python
def scatter1():

    ax = data.plot(x="Weight", y="Height", kind='scatter')

    ax.set_ylabel(u", ")

    ax.set_xlabel(u", ")

    ax.set_title(u"   ")

    return ax

scatter1()
```

```python
f_err = lambda w0, w1: sum((data['Height'] - (w0 + w1*data['Weight'])) ** 2)
```

```python

def plot_line(w0, w1, c='r'):

    x = np.linspace(min(data['Weight']), max(data['Weight']), 100)

    plt.plot(x, w0 + w1 * x, c)

# plt.scatter(data['Height'], data['Weight'])

scatter1()

plot_line(60, 0.05)

plot_line(50, 0.16, 'g')

plt.show()
```

```python

e50 = lambda w1: f_err(50, w1)

x = np.linspace(-0.2, 0.4, 100)

y = [e50(w1) for w1 in x]

plt.plot(x, y)

plt.xlabel(u' $w_1$')

plt.ylabel(u'')

plt.title(u'    $w_1$  $w_0$ = 50')
```

```python
from scipy.optimize import minimize_scalar

w1_opt = minimize_scalar(e50, bounds=(-5, 5), method='bounded').x

w1_opt
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()

ax = fig.gca(projection='3d') # get current axis
```

```python
X = np.arange(-5, 5, 0.25)

Y = np.arange(-5, 5, 0.25)

X, Y = np.meshgrid(X, Y)

Z = np.sin(np.sqrt(X**2 + Y**2))
```

```python
surf = ax.plot_surface(X, Y, Z)

ax.set_xlabel('X')

ax.set_ylabel('Y')

ax.set_zlabel('Z')

plt.show()

```

```python

from matplotlib import cm
```

```python
step = 40

w0 = np.linspace(40, 80, step)

w1 = np.linspace(-0.5, 0.5, step)

xs, ys = np.meshgrid(w1, w0)

zs = np.zeros((len(w0), len(w1)))

for i in range(len(w0)):

    for j in range(len(w1)):

        zs[i, j] = f_err(w0[i], w1[j])

fig = plt.figure()

ax = fig.gca(projection='3d')

surf = ax.plot_surface(xs, ys, zs, rstride=1, cstride=1, linewidth=0, cmap=cm.coolwarm, antialiased=False)

ax.set_ylabel('Intercept')

ax.set_xlabel('Slope')

ax.set_zlabel('Error')

plt.show()
```

```python
from scipy.optimize import minimize

v_err = lambda v: f_err(v[0], v[1])

w_m = minimize(v_err, [0,0], method='L-BFGS-B', bounds=[(-100,100),(-5,5)]).x

print (w_m[0], w_m[1])
```

```python
scatter1()

plot_line(w_m[0],w_m[1])

plt.show()
```
